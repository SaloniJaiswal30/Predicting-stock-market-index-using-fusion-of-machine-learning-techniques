```python
from tensorflow.python.keras.models import Sequential
from keras import optimizers
from tensorflow.python.keras.layers import Dense, Activation
import pandas as pd
```

```python
data = pd.read_csv('paramdataset.csv', header=None)
```

```python
from sklearn.model_selection import train_test_split
#train_x=data.iloc[:2452,1:11]
#train_y=data.iloc[:2452,12]
#test_x=data.iloc[2452:2455,1:11]
#test_y=data.iloc[2452:2455,12]
train_x, test_x, train_y, test_y = train_test_split(data.iloc[:,1:11], data.iloc[:,12], test_size=0.2, random_state=None, shuffle=False)
```

```python
#print(test_y)
```

```python
def svr_model(train_x, train_y, test_x):
    from sklearn.svm import SVR
    import numpy as np

    clf = SVR(C=1.0, epsilon=2.0, gamma=1e-8,verbose=False)
    clf.fit(train_x, train_y)
    predicted_y = clf.predict(test_x)
    print(predicted_y)
    return predicted_y
```

```python
predict_svr=svr_model(train_x, train_y, test_x)
#print(predict_svr)
```

```python
def random_forest_reg(train_x, train_y, test_x):
    from sklearn.ensemble import RandomForestRegressor
    import numpy as np
    rfr = RandomForestRegressor(max_depth=2, random_state=0)
    rfr.fit(train_x, train_y)
    rfr_predicted_y = rfr.predict(test_x)
    return rfr_predicted_y
```

```python
predict_rf=random_forest_reg(train_x, train_y, test_x)
#print(predict_rf)
```

```python
def ann_model(train_x, train_y, test_x):
    model = Sequential()
    model.add(Dense(64, activation='sigmoid', kernel_initializer='normal', input_dim=10))
    model.add(Dense(1,activation='linear', kernel_initializer='normal'))
    #sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)

    model.compile(optimizer='sgd', loss = 'mean_squared_error', metrics = ['accuracy'])

    model.fit(train_x, train_y,
              epochs=300,
              batch_size=10)
    ann_predicted_y = model.predict(test_x)
    ann_predicted_y = ann_predicted_y.reshape(-1)
    return ann_predicted_y
```

```python
predict_ann = ann_model(train_x, train_y, test_x)
#print(predict_ann)
#print(test_y)
```

```python
def error_function(test_y, predicted_y):
    import numpy as np
    from math import sqrt
    test = list(test_y)
    MAPE=[(test[i]-predicted_y[i])/test[i] for i in range(len(test))]
    MAPE1=abs((sum(MAPE)*100/len(test)))
    MAE=abs((sum(MAPE)/len(test)))
    rRMSE=[((test[i]-predicted_y[i])/test[i])**2 for i in range(len(test))]
    rRMSE=sqrt(sum(rRMSE)/len(test))
    MSE=[(test[i]-predicted_y[i])**2 for i in range(len(test))]
    MSE=sum(MSE)/len(test)
    #print(MSE)
    return MAPE1,MAE,rRMSE,MSE
```

```python
print("\t MAPE\t\t\tMAE\t\t\trRMSE\t\t\tMSE")
print("SVR:",error_function(test_y, predict_svr))
print("RF:",error_function(test_y, predict_rf))
print("ANN:",error_function(test_y, predict_ann))
```
